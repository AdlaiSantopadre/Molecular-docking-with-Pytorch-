```python
# Import delle librerie
import selfies as sf
from IPython.display import display, Markdown
```

```python
# Inserisci la tua stringa SMILES
smiles_input = "CCCCCC1C2CCCC1CCCC2N1CNC(C)(C)C1C(C)(C)CCCC1CFC"
```

```python
# Conversione SMILES â†’ SELFIES â†’ SMILES

selfies_str = sf.encoder(smiles_input)

decoded_smiles = sf.decoder(selfies_str)
```

```python
# Visualizzazione dei risultati
display(Markdown(f"**SMILES originale:** `{smiles_input}`"))
display(Markdown(f"**SELFIES:** `{selfies_str}`"))
display(Markdown(f"**SMILES decodificato da SELFIES:** `{decoded_smiles}`"))
```

```python
# Generazione link per MolView
molview_url = f"https://molview.org/?smiles={decoded_smiles}"
display(Markdown(f"### ðŸ”¬ [Visualizza su MolView in 3D]({molview_url})"))
```
